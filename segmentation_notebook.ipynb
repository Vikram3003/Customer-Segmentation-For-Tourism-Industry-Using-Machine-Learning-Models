```python
#Implementing K-means Algorithm
```

```python
import numpy as np
from scipy import stats
```

```python
new_df = df[['Age','Avg. expense of touring each year (Rs.)','Avg. money saved each year for touring (Rs.)']]

#Remove Outliers
z_scores = stats.zscore(new_df)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
new_df = new_df[filtered_entries]
```

```python
new_df.info()
```

```python
from sklearn.cluster import KMeans
```

```python
SSE = []
for cluster in range(1,10):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(new_df)
    SSE.append(kmeans.inertia_)
    
# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,10), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# First, build a model with 4 clusters
kmeans = KMeans( n_clusters = 4, init='k-means++')
kmeans.fit(new_df)
```

```python
from sklearn.metrics import silhouette_score
```

```python
print(silhouette_score(new_df, kmeans.labels_, metric='euclidean'))
```

```python
pred = kmeans.predict(new_df)
frame = pd.DataFrame(new_df)
frame['cluster'] = pred
```

```python
frame.head()
```

```python
list1 = ['Age','Avg. expense of touring each year (Rs.)','Avg. money saved each year for touring (Rs.)']
avg_df = frame.groupby(['cluster'], as_index=False).mean()
for i in list1:
    sns.barplot(x='cluster',y=str(i),data=avg_df)
    plt.show()
```

```python
#Implementing Affinity Propagation Algorithm
```

```python
from sklearn.cluster import AffinityPropagation
from sklearn import metrics
from itertools import cycle
from sklearn.datasets import make_blobs
```

```python
#Affinity Propagation for Age Vs Avg. expense of touring each year (Rs.)
centers = df[['Age', 'Avg. expense of touring each year (Rs.)']]
```

```python
X, labels_true = make_blobs(n_samples = 2660, centers = centers, cluster_std = 0.5, random_state = 0)
```

```python
# Compute Affinity Propagation
af = AffinityPropagation(preference =-50).fit(X)
cluster_centers_indices = af.cluster_centers_indices_
labels = af.labels_
  
n_clusters_ = len(cluster_centers_indices)
```

```python
print(n_clusters_)
```

```python
colors = cycle('bgrcmykbgrcmykbgrcmykbgrcmyk')
  
for k, col in zip(range(n_clusters_), colors):
    class_members = labels == k
    cluster_center = X[cluster_centers_indices[k]]
    plt.plot(X[class_members, 0], X[class_members, 1], col + '.')
    plt.plot(cluster_center[0], cluster_center[1], 'o',
             markerfacecolor = col, markeredgecolor ='k',
             markersize = 14)
  
    for x in X[class_members]:
        plt.plot([cluster_center[0], x[0]], 
                 [cluster_center[1], x[1]], col)
  
plt.title('Estimated number of clusters: % d' % n_clusters_)
plt.show()
```
